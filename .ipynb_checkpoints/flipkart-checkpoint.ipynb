```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import time



productname = []
price = []
ratings = []
features = []
pagenum = []
original_price = []
Discount = []
Review = []
headers = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/147.0.0.0 Safari/537.36"}

total_time = time.time()

for i in range(1,93):
    start_time = time.time()

    URL = f"https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&as-pos=1&as-type=RECENT&suggestionId=laptop%7CLaptops&requestId=3f3ea89d-e660-48a9-b1a4-5c9c322da66e&as-searchtext=page{i}"
    page = requests.get(url,headers = headers)
    soup = BeautifulSoup(page.text, "html.parser")
    for x in soup.find_all("div", attrs={"class":"ZFwe0M row"}):
        pname = x.find("div", attrs={"class":"RG5Slk"})
        cost = x.find("div", attrs={"class":"hZ3P6w DeU9vF"})
        rat = x.find("div", attrs={"class":"MKiFS6"})
        specs = x.find("div", attrs={"class":"CMXw7N"})
        op = x.find("div", attrs={"class":"kRYCnD gxR4EY"})
        discount = x.find("div", attrs={"class":"HQe8jr"})
        reviews = x.find("span", attrs={"class":"PvbNMB"})

        productname.append(pname.text if pname else np.nan)
        price.append(cost.text if cost else np.nan)
        ratings.append(rat.text if rat else np.nan)
        features.append(specs.text if specs else np.nan)
        original_price.append(op.text if op else np.nan)
        Discount.append(discount.text if discount else np.nan)
        Review.append(reviews.text if reviews else np.nan)

        pagenum.append(i)
    print(f"page {i} completed in {time.time()-start_time:.2f} seconds")
print("Total Time Completed in seconds", str(time.time()-total_time))
        
```

```python
df = pd.DataFrame({
    "Product Name": productname,
    "Price": price,
    "Original Price": original_price,
    "Discount": Discount,
    "Ratings": ratings,
    "Reviews": Review,
    "Features":features,
    "Page Number": pagenum
    
})
```

# 2. Bivariate Analysis

# A. Continuous vs Continuous

# Price vs Ratings

```python
sns.scatterplot(x='Price', y='ratings', data=df)
plt.show()
```

# Interpretation:
- Expensive laptops do not always receive higher ratings.
- Weak positive relationship between price and rating.

# RAM vs Price

```python
sns.scatterplot(x='RAM', y='Price', data=df)
plt.show()
```

# Interpretation:
- Higher RAM configurations generally have higher prices.

# Storage vs Price

```python
sns.scatterplot(x='Storage', y='Price', data=df)
plt.show()
```

# Interpretation:
- Laptops with larger storage tend to be priced higher.

# Correlation Heatmap

```python
sns.heatmap(df.corr(numeric_only = True),
           annot = True,
           cmap = 'coolwarm')
plt.show()
```

# Interpretation:
- Price is positively correlated with RAM and Storage.
- Ratings have relatively weak correlations with technical specifications.

# B. Categorical vs Continuous

# Brand vs Price

```python
sns.boxplot(x='Brand', y='Price', data=df)
plt.xticks(rotation=90)
plt.show()
```

# Interpretation:
- Apple laptops have higher median prices.
- Budget brands show lower price distributions.

# Brand vs Ratings

```python
sns.boxplot(x='Brand', y='ratings', data=df)
plt.xticks(rotation=90)
plt.show()
```

# 4. Final Conclusion
1. HP and Apple have the largest number of laptop listings.
2. Intel processors dominate the laptop market.
3. Most laptops come with 16 GB RAM and 512 GB storage.
4. Laptop prices are right-skewed due to premium models.
5. Higher RAM and Storage generally increase laptop prices.
6. Apple laptops have the highest average prices.
7. Ratings are mostly between 4.0 and 4.5, indicating good customer satisfaction.
8. Price has only a weak relationship with ratings.
9. Intel and AMD processors cover a wide range of price segments, while Apple processors are mostly premium.
10. Premium-priced laptops are not necessarily rated significantly higher than budget laptops.
